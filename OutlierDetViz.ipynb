# **DETECTING OUTLIERS USING VISUALISATION AND GRAPHS**

Data visualization is useful for data cleaning, exploring data, detecting outliers and unusual groups, identifying trends and clusters etc. Here the list of data visualization plots to spot the outliers.



---



1.   BOX AND WHISKER PLOT
2.   SCATTER PLOT
3.   HISTOGRAM 
4.   DISTRIBUTION PLOT
5.  QQ PLOT


```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from statsmodels.graphics.gofplots import qqplot
train = pd.read_csv('/content/bhp.csv')
train.head()
```

```python

def Box_plots(df):
    plt.figure(figsize=(10, 4))
    plt.title("Box Plot")
    sns.boxplot(df)
    plt.show()
Box_plots(train['price_per_sqft'])


def scatter_plots(df1,df2):
    fig, ax = plt.subplots(figsize=(10,4))
    ax.scatter(df1,df2)
    ax.set_xlabel('total_sqft')
    ax.set_ylabel('price')
    plt.title("Scatter Plot")
    plt.show()
scatter_plots(train['total_sqft'],train['price'])

def dist_plots(df):
    plt.figure(figsize=(10, 4))
    sns.distplot(df)
    plt.title("Distribution plot")
    sns.despine()
    plt.show()
dist_plots(train['price'])

def qq_plots(df):
    plt.figure(figsize=(10, 4))
    qqplot(df,line='s')
    plt.title("Normal QQPlot")
    plt.show()
qq_plots(train['price'])
```
